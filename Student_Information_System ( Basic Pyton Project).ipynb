```python

student_info = {}     # use an empty dictionary to store informaton

while True:
    print("\n===== Student Information System =====")
    print("1. Add Student")
    print("2. View Student Information")
    print("3. Update Student Information")
    print("4. Delete Student")
    print("5. Exit")

    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '1':
        student_id = input("Enter Student ID: ")
        name = input("Enter Student Name: ")
        age = input("Enter Student Age: ")
        grade = input("Enter Student Grade: ")

        student_info[student_id] = {
            'Name': name,
            'Age': age,
            'Grade': grade
        }
        print(f"Student with ID {student_id} added to the system.")

    elif choice == '2':
        student_id = input("Enter Student ID to view information: ")
        if student_id in student_info:
            print("\n===== Student Information =====")
            print("Student ID:", student_id)
            for key, value in student_info[student_id].items():
                print(f"{key}: {value}")
        else:
            print(f"Student with ID {student_id} not found in the system.")

    elif choice == '3':
        student_id = input("Enter Student ID to update information: ")
        if student_id in student_info:
            name = input("Enter new Student Name: ")
            age = input("Enter new Student Age: ")
            grade = input("Enter new Student Grade: ")

            student_info[student_id] = {
                'Name': name,
                'Age': age,
                'Grade': grade
            }
            print(f"Student information for ID {student_id} updated.")
        else:
            print(f"Student with ID {student_id} not found in the system.")

    elif choice == '4':
        student_id = input("Enter Student ID to delete: ")
        if student_id in student_info:
            del student_info[student_id]
            print(f"Student with ID {student_id} deleted from the system.")
        else:
            print(f"Student with ID {student_id} not found in the system.")

    elif choice == '5':
        print("Exiting Student Information System.")
        break

    else:
        print("Invalid choice. Please choose 1, 2, 3, 4, or 5.")
```
