#Matplotlib_plotting

```python
# Import  Libraries
import numpy as np
import matplotlib.pyplot as plt

# Body

n = np.linspace(0, 4*np.pi, 100)
sin = np.sin(n)
plt.plot(n, sin)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Sin Wave")
plt.xlim([0, 4*np.pi])
plt.ylim(-1.5, 1.5)
plt.show()
```

#Matrix

```python
import numpy as np
from scipy.linalg import lu
a = np.mat('1, 2; 3, 4; 5, 6')
print(a)
```

```python
aT = np.transpose(a)
print(aT)

```

```python
b = 2*np.eye(3)
print(b)
```

```python
c = b*a
print(c)
```

```python
l = lu(a)
print(l)
```
